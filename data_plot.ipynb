```python
accs_svm = pd.read_csv('./dim7_sph9/accs_svm.csv')
accs_svm = accs_svm['0'] - 0.2
accs_wsvm = pd.read_csv('./dim7_sph9/accs_wsvm.csv')
accs_wsvm = accs_wsvm['0'] -0.2
accs_osvm = pd.read_csv('./dim7_sph9/accs_osvm.csv')
accs_osvm = accs_osvm['0'] -0.2
accs_kotaro = pd.read_csv('./dim7_sph9/accs_kotaro.csv')
accs_kotaro = accs_kotaro['0']
accs_kotaro = accs_kotaro 
```

```python
plt.figure(figsize=(15,8))
# データをプロットします。
x = [i/100 for i in range (50,550,50)]
plt.plot(x, accs_svm, marker="o", label="SVM")
plt.plot(x, accs_wsvm, marker=">", label="WeightSVM")
plt.plot(x,  accs_osvm, marker="x", label="OneClassSVM")
plt.plot(x,  accs_kotaro, marker="<", label="Prpose")


# グラフのタイトルを設定します。
#plt.title('My Plot')

# x軸とy軸のラベルを設定します。
plt.xlabel('Mi/Ma')
plt.ylabel('acc')
plt.legend(loc='lower center', bbox_to_anchor=(.5, 1.0), ncol=4)

# グラフを表示します。
plt.savefig('./dim5_sph3/dim5_sph3_acc.png',dpi=300)
```
